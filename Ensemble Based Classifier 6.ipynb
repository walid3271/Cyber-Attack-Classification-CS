```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# Logistic
model_logistic = LogisticRegression()
model_logistic.fit(x_train, y_train)

y_pred_logistic = model_logistic.decision_function(x_test)
```

```python
# DecisionTree
model_decision = DecisionTreeClassifier()
model_decision.fit(x_train, y_train)

y_pred_decision = model_decision.predict_proba(x_test)[:, 1]
```

```python
# Naive Bayes
model_gaussian = GaussianNB()
model_gaussian.fit(x_train, y_train)

y_pred_gaussian = model_gaussian.predict_proba(x_test)[:, 1]
```

```python
from sklearn.metrics import roc_curve, auc

logistic_fpr, logistic_tpr, threshold = roc_curve(y_test, y_pred_logistic)
auc_logistic = auc(logistic_fpr, logistic_tpr)

decision_fpr, decision_tpr, threshold = roc_curve(y_test, y_pred_decision)
auc_decision = auc(decision_fpr, decision_tpr)

gaussian_fpr, gaussian_tpr, threshold = roc_curve(y_test, y_pred_gaussian)
auc_gaussian = auc(gaussian_fpr, gaussian_tpr)

plt.figure(figsize=(5, 5), dpi=100)

plt.plot(logistic_fpr, logistic_tpr, marker='.', label='Logistic (auc = %0.3f)' % auc_logistic)

plt.plot(decision_fpr, decision_tpr, marker='.', label='Decision (auc = %0.3f)' % auc_decision)

plt.plot(gaussian_fpr, gaussian_tpr, marker='.', label='Gaussian (auc = %0.3f)' % auc_gaussian)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')
plt.title('ROC CURVE')

plt.legend()

plt.show()
```

```python
clf1 = GaussianNB()
clf2 = DecisionTreeClassifier()
clf3 = LogisticRegression()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
# Initialize a list to store the individual voting accuracy for each fold of Naive Bayes.
fold_accuracies = []

# Perform k-fold cross-validation
k = 10  # Set the desired number of folds
skf = StratifiedKFold(n_splits=k)

for train_index, test_index in skf.split(x_train, y_train):
    x_train_fold, x_test_fold = x.iloc[train_index], x.iloc[test_index]
    y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]

    # Fit the classifier on the training fold
    clf1.fit(x_train_fold, y_train_fold)

    # Make predictions on the test fold using individual voting
    predictions = clf1.predict(x_test_fold)

    # Calculate accuracy for the fold and append to the list
    fold_accuracy = accuracy_score(y_test_fold, predictions)
    fold_accuracies.append(fold_accuracy)

# Print the accuracy for each fold
for fold, accuracy in enumerate(fold_accuracies):
    print(f"Accuracy for Fold {fold+1}: {accuracy}")
    
# Create a bar plot of the fold accuracies
plt.figure(figsize=(8, 6))
plt.bar(range(1, k+1), fold_accuracies)
plt.xlabel('Accuracy for each fold')
plt.ylabel('Accuracy')
plt.title('10 Fold of Naive Bayes classifier')
plt.show()
```
