```python
y=ds['0.33']
```

```python
# replace the infinite number with a biggest finite number.
# inplace=True --> Do this replacement directly in the list.
x.replace([np.inf, -np.inf], np.finfo(np.float64).max, inplace=True)
```

```python
from sklearn.feature_selection import RFE
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
```

```python
# convert to 0 and 1 all data of the dataset
scaler = MinMaxScaler() 
x_scaled = scaler.fit_transform(x)
```

```python
estimator = LogisticRegression()
```

```python
n_features_to_select = 25
```

```python
#Recursive Feature Elimination (RFE)
rfe = RFE(estimator)
rfe.n_features_to_select = n_features_to_select
x_scaled = rfe.fit_transform(x, y)

# All selected feature column is stored.
selected_feature_names = x.columns[rfe.support_] 
print(selected_feature_names)
```

```python
x = ds[selected_feature_names]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
classifiers= [['Naive Bayes:',GaussianNB()],['LogisticRegression:',LogisticRegression(max_iter=1000)],['DecisionTree:',DecisionTreeClassifier()]]
```

```python
# used to store the predictions made by each classifier.
predictions_ds=pd.DataFrame() 

# Adding a Column to the DataFrame
predictions_ds['action']=y_test 
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Iterate over classifiers
for name, classifier in classifiers:
    classifier.fit(x_train, y_train.ravel())
    predictions = classifier.predict(x_test)
    # store the predictions for each classifier in a dictionary, where the key is the name of the classifier without the trailing colon (':').
    predictions_ds[name.strip(":")] = predictions 

    # Calculate evaluation scores
    accuracy = accuracy_score(y_test, predictions)
    # calculates the score for each label and then takes the unweighted mean of the scores.
    # to get a balanced view of the model's performance across all classes
    recall = recall_score(y_test, predictions, average='macro')  
    precision = precision_score(y_test, predictions, average='macro')
    f1 = f1_score(y_test, predictions, average='macro')

    # Print evaluation scores
    print(f"Metrics for {name}")
    print(f"Accuracy : {accuracy}")
    print(f"Recall   : {recall}")
    print(f"Precision: {precision}")
    print(f"F1 Score : {f1}")
    print("------------------------")
```

```python
# Logistic
model_logistic = LogisticRegression()
model_logistic.fit(x_train, y_train)

# [:, 1] extract specific column from 2d matrix.
# ":," before comma means want to include all rows.
# ",1" after the comma means to select second column.
y_pred_logistic = model_logistic.predict_proba(x_test)[:, 1]
```

```python
from sklearn.metrics import roc_curve, auc


logistic_fpr, logistic_tpr, threshold = roc_curve(y_test, y_pred_logistic)
# AUC is a performance metric that measures the overall quality of the classifier's predictions. It ranges from 0 to 1, with a higher value indicating better performance.
auc_logistic = auc(logistic_fpr, logistic_tpr) 

decision_fpr, decision_tpr, threshold = roc_curve(y_test, y_pred_decision)
auc_decision = auc(decision_fpr, decision_tpr)

gaussian_fpr, gaussian_tpr, threshold = roc_curve(y_test, y_pred_gaussian)
auc_gaussian = auc(gaussian_fpr, gaussian_tpr)

plt.figure(figsize=(5, 5), dpi=100)

plt.plot(logistic_fpr, logistic_tpr, marker='.', label='Logistic (auc = %0.3f)' % auc_logistic)

plt.plot(decision_fpr, decision_tpr, marker='.', label='Decision (auc = %0.3f)' % auc_decision)

plt.plot(gaussian_fpr, gaussian_tpr, marker='.', label='Gaussian (auc = %0.3f)' % auc_gaussian)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')
plt.title('ROC CURVE')

plt.legend()

plt.show()
```

```python
clf1 = GaussianNB()
clf2 = DecisionTreeClassifier()
clf3 = LogisticRegression()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

k = 5
c = []

classifiers = [clf1, clf2, clf3]
# Loop through each classifier
for clf in classifiers:
    accuracy_scores = cross_val_score(clf, x_train, y_train, cv=k)
    c.append(accuracy_scores)

    # Print the accuracy of each fold for the current classifier
    print(f"Classifier: {clf.__class__.__name__}")
    for fold_number, accuracy in enumerate(accuracy_scores, 1):
        print(f"Fold {fold_number}: Accuracy = {accuracy:.2f}")
    print()
    
    
# List of fold numbers (1 to k) for the x-axis
fold_numbers = np.arange(1, k+1)

colors = ['r', 'g', 'b']

# Plotting the bar plot for each classifier
for i, clf in enumerate(classifiers):
    accuracy_scores = c[i]
    
    # Plot the bar graph for the current classifier
    plt.bar(fold_numbers + i*0.2, accuracy_scores, width=0.2, color=colors[i], label=clf.__class__.__name__)

# Add labels and title to the plot
plt.xlabel('Fold Number')
plt.ylabel('Accuracy')
plt.title('Accuracy for each fold and classifier')
plt.xticks(fold_numbers + 0.2, fold_numbers)
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier

# estimators: This is a list of tuples, where each tuple consists of a name and a classifier. # majority voting.
eclf1 = VotingClassifier(estimators = [('Gaussian',clf1), ('DecisionTree',clf2), ('Logistic',clf3)], voting='hard') 
eclf1.fit(x_train,y_train)
#print("Accuracy after hard voting is: ",eclf1.score(x_test,y_test))


y_pred = eclf1.predict(x_test)
# calculates the score for each label and then takes the unweighted mean of the scores. 
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro') 
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

# Print evaluation scores
print("After hard voting.")
print(f"Accuracy : {accuracy:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"Precision: {precision:.4f}")
print(f"F1 Score : {f1:.4f}")
```

```python
# hyperparameters for the individual classifiers
# Ensemble classifier is configured using the hyperparameters for three different individual classifiers
# C is the hyperparameter that represents the inverse of the regularization strength.
param_grid = {
    # randomly search these values to find the best combination of hyperparameters.
    'Logistic__C': [0.1, 1.0, 10.0], 
    # maximum depth or levels of the decision tree. None means the tree can grow deeper and potentially capture more complex relationships in the data.
    # limits the depth of the Decision Tree to a maximum of 5 levels. 
    # Maximum of 10 levels, when set to 5 but still constrains its complexity.
    'DecisionTree__max_depth': [None, 5, 10],
    # adds a small value to the variances of the features during calculations.
    # Handle if feature's variance is zero or very close to zero.
    'Gaussian__var_smoothing': [1e-09, 1e-08, 1e-07],  
    'voting': ['hard', 'soft']
}
```

```python
# Perform grid search with cross-validation to find the best hyperparameters
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(eclf1, param_grid, cv=5)
grid_search.fit(x_train, y_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
```

```python
# Update the individual classifiers with the best hyperparameters
# ** the key-value pairs inside the dictionary will be treated as keyword arguments.
clf1.set_params(**{'var_smoothing': best_params['Gaussian__var_smoothing']})
clf2.set_params(**{'max_depth': best_params['DecisionTree__max_depth']})
clf3.set_params(**{'C': best_params['Logistic__C']})
```

```python
# Create the ensemble classifier with the best hyperparameters
eclf1 = VotingClassifier(estimators = [('Gaussian',clf1), ('DecisionTree',clf2), ('Logistic',clf3)], voting=best_params['voting'])
```

```python
# Train the ensemble classifier on the full training set
eclf1.fit(x_train, y_train)
```

```python
# Evaluate the ensemble classifier on the test set
y_pred = eclf1.predict(x_test)

# calculates the score for each label and then takes the unweighted mean of the scores.
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')  
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

# Print evaluation scores
print("After hyperparameters tuning.")
print(f"Accuracy : {accuracy:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"Precision: {precision:.4f}")
print(f"F1 Score : {f1:.4f}")
```
